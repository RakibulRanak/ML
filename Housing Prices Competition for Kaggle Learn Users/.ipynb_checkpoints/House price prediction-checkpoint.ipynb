```python
import matplotlib.pyplot as plt
import seaborn as sns
train.isnull()
for i in range(1,5):
    plt.figure(figsize=(21, 6))
    temp  = train.iloc[:,(i-1)*21:i*21]
    sns.heatmap(temp.isnull())
    plt.show()
```

```python
train.drop(['Alley'], axis=1, inplace=True)
train.drop(['FireplaceQu'], axis=1, inplace=True)
train.drop(['PoolQC'], axis=1, inplace=True)
train.drop(['Fence'], axis=1, inplace=True)
train.drop(['MiscFeature'], axis=1, inplace=True)
```

```python
import numpy as np
data_corr = train.select_dtypes(include=[np.number])
#data_corr = train
data_corr.fillna(0, inplace=True)
data_corr.head()
```

```python
import seaborn as sns
corr = data_corr.corr()
sns.heatmap(corr, cmap = 'coolwarm',  annot=True)
```

```python
corr.sort_values(['SalePrice'], ascending=False, inplace=True)
corr['SalePrice']
```

```python
import numpy as np
data_corr2 = train.select_dtypes(exclude=[np.number])
#data_corr = train
data_corr2.fillna('AAA', inplace=True)
data_corr2.head()
```

```python
#data_corr2 = preprocessing.LabelEncoder().fit_transform(data_corr2.values)
data_corr2 = data_corr2.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
data_corr2['SalePrice']=train['SalePrice']
print(data_corr2.head())
corr2 = data_corr2.corr()
sns.heatmap(corr2, cmap = 'coolwarm',  annot=True)
corr2.sort_values(['SalePrice'], ascending=False, inplace=True)
corr2['SalePrice']
```
